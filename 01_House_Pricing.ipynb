```python
results = []
```

```python
# Iterate over test data
for index, row in test_df1.iterrows():
    # Reshape row to (1, -1) because model expects 2D input
    input_data = row.values.reshape(1, -1)
    
    # Predict using the model
    prediction = model.predict(input_data)
    
    # Append result as a dictionary
    results.append({
        'Index': index,
        'FullBath': row['FullBath'],
        'HalfBath': row['HalfBath'],
        'BedroomAbvGr': row['BedroomAbvGr'],
        'LotArea': row['LotArea'],
        'Predicted': (prediction)
    })
    
    # Convert to DataFrame and show (optional: round to 2 decimals)
    iter_df = pd.DataFrame(results)
    print(iter_df.round(2))
    print('-' * 80)
    break
```
